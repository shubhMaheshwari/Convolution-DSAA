# Applying and implementing convolution

```matlab
img = imread('./cameraman.png');
imshow(img)
size(img);
```

```matlab
M = [1 2 1; 0 0 0; -1 -2 -1];
M = double(M);
```

# Detecting Horizontal Edges

```matlab
new_img = conv2(M,img);
new_img = mat2gray(new_img);
imshow(new_img)
```

# Detecting Vertical Edges
By taking the transpose we can find Vertical images

```matlab
new_img = conv2(M',img);
new_img = mat2gray(new_img);
imshow(new_img)
```

Originally we can't see much difference but after converting it to 
gray we can see the edges detected
