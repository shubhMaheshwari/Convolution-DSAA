# 2 D Convolution and edge detection

Let assume the below 3x3 matrix

```matlab
M = [1,2,1;0,0,0;-1,-2,-1]
```

We can see after convolution, we get a red line.

```matlab
img = imread('./sample_inp.png');
new_img = myconv(M,img);
```

```matlab
imshow(new_img);
```

### Using the M we first detect many vertical edges

```matlab
blur = imread('./blur.jpg');
ver_blur = myconv(M,blur);
imshow(ver_blur*5)
```

### Using the M' we detect many horizontal edges

```matlab
hor_blur = myconv(M',blur);
imshow(hor_blur*5)
```

```matlab
fin_blur = ver_blur + hor_blur;
imshow(fin_blur.*5)
```

Hence we can even do this on other images`

```matlab
img = imread('./Faces.jpg');
ver_img = myconv(M,img);
hor_img = myconv(M',img);
fin_img = ver_img + hor_img;
imshow(fin_img.*5)
```

```matlab
img = imread('./War_on_drugs.png');
ver_img = myconv(M,img);
hor_img = myconv(M',img);
fin_img = ver_img + hor_img;
imshow(fin_img.*5)
```

###  We get better edge detection after using some kind of blur

```matlab
fin_img = conv2([1,1,1;1,1,1;1,1,1]./9,fin_img);
imshow(fin_img)
```
